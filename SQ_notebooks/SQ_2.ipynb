The special thing about atan2 as opposed to atan is that the former will yield a result between the full range of possible angles, ie from -180° to 180°. The latter only goes from -90° to 90°. This means that atan2 will give results for any possible point on a graph, and atan will only yield results for points in the positive x quadrants.
