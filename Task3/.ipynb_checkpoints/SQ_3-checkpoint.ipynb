```python
##Question 3.1 Session quiz
masses = [1.9891e+30, 1.8986e+27, 
          5.6846e+26, 1.0243e+26, 8.6810e+25,
          5.9736e+24, 4.8685e+24, 6.4185e+23, 
          3.3022e+23, 7.349e+22, 1.25e22]

Slasses = slice(6, 11)
Massies = masses[Slasses] #sliced list of only the last 5

MM=7.349e+22 #mass of the moon
for M in masses:
    if M/MM < 1:
        masses.remove(M)
for M in masses:
    if M/MM == 1:
        masses.remove(M)
print(masses)
print()
print(Massies)

ObiMass = sum(Massies) #total mass of the new list
Lenght= len(Massies) #new lists length
Amass = ObiMass / Lenght #average mass
print(Amass)
```

```python
##Question 3.2 Session quiz
import math
from scipy.constants import G
masses = [1.9891e+30, 1.8986e+27, 5.6846e+26, 
          1.0243e+26, 8.6810e+25, 5.9736e+24, 
          4.8685e+24, 6.4185e+23, 3.3022e+23, 
          7.349e+22, 1.25e22]
radii  = [696340, 71492, 60268, 24764, 
          25559, 6378.1, 6051.8, 3396.2, 
          2439.7, 1738.1, 1195]
names  = ['Sun', 'Jupiter', 'Saturn', 'Neptune', 
          'Uranus', 'Earth', 'Venus', 'Mars', 
          'Mercury', 'Moon', 'Pluto']


for N, M, R in zip(names, masses, radii):
    V = math.sqrt((2*G*M)/R)
    print(N,"=", V, "m/s", 'with a mass of', f'{M:3f}', 'kg')

bomboclat = {}
    
for N, M, R in zip(names, masses, radii):
    V = math.sqrt((2*G*M)/R)
    bomboclat[N] = V
print(bomboclat)
```

```python
##3.7 Session quiz
import numpy
d=float(input('Size: '))
D=float(input('Distance: '))
def delta(d, D):
    δ = 2*numpy.arctan((d)/(2*D))
    return δ

δd = numpy.degrees(delta(d, D))
print(f'The angular size δ = {δd} degrees')
```

```python
#3.10 Session quiz
import math
#r = input('input radius')
#h = input('input height')
##print(Vc)

def cylinder(R, H):
    return(2*(R**2)*math.pi*H +2*R*math.pi*H, math.pi*R**2*H)
R = float(input('Input radius: '))
H = float(input('Input height: '))

a, v = cylinder(R, H)
print(f'The cylinders area A= {a}')
print(f'The cylinders volume V= {v}')
print(f'The ratio are to volume: {a/v}')
```

```python
##Question 3.18 Session quiz
V = np.array([2094.75, 2320.87, 1224.9, 872.13, 432.2, 751, 1410, 2700, 5292, 2550, 5253, 2419, 2158])
D = np.array([18.49, 35.48, 35.48, 11.34, 7.12, 9.82, 23.66, 44, 66.7, 52.4, 55, 25.6, 26.19])
# y = V and x = D

xbar= np.mean(D)
ybar= np.mean(V)
N=0
S=0
for i in range(len(D)):
    N+=(D[i]-xbar)*(V[i]-ybar)
    S+= (D[i]-xbar)**2
b=N/S
a=ybar-b*xbar
print(f"The slope b = {b}")
print('The intercept a =', a)
```
