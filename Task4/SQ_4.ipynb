```python
#question 4.4
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
NH= np.random.normal(185, 8, 100000)

NHm = np.mean(NH)
NHstd = np.std(NH)
print(f'The mean of the randomly generated distribution = {NHm}')
print(f'The standard deviation of the randomly generated distribution = {NHstd}')
print('These values are quite similar to the given values of the mean and std when generating the distribution')

plt.title('The intensity of all the average heights (cm)')
plt.hist(NH, bins=40, color='blue', label='Average height intensity')
plt.plot([], [], color='white', label=f'The mean = {NHm:.4f}')
plt.plot([], [], color='white', label=f'The std = {NHstd:.4f}')
plt.axvline(NHm, color='green', linestyle='--', label='sample mean', linewidth=3, alpha = 0.5)
plt.axvspan(NHm - NHstd, NHm + NHstd, label='std span', alpha=0.2, color='red')
plt.xlabel('Average height (cm)')
plt.ylabel('Intensity')
plt.legend()
plt.show()

h=190
P = norm.cdf(h, loc=185, scale=8)
print(f'The probability of someone having a height =< h = {P}')
top = 185+8
bot = 185-8

Pr = norm.cdf(top, loc=185, scale=8)-norm.cdf(bot, loc=185, scale=8)
print(f'Thus about {(Pr*100):.3f}% of values lie within one standard deviation')
```

```python
#Question 4.12
import numpy as np
import matplotlib.pyplot as plt
x = np.array([1,3,3,1,1])
y = np.array([1,1,3,3,1])
z= x+1j*y
deg = 30
rad = np.deg2rad(deg)
rotot = np.exp(1j*rad)
Zr = z*rotot
Xr = Zr.real #takes the real part of the complex
Yr = Zr.imag #takes the complex/imaginary part

plt.plot(x, y, label='OG square')
plt.plot(Xr, Yr, label=f'Rotated 30 degrees')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#question 4.14
import numpy as np
def funcfilter(func, seq):
    S = np.asarray(seq)
    NZ=S[S>0]
    return func(NZ)
sq = np.array([-100, -10, 0, 10, 100])
result = funcfilter(np.log10, sq)
print("Result log10:", result)
resultS=funcfilter(np.log10, 100)
print("result log10 single:", resultS)
```

```python
#Question 4.16
import numpy as np
import scipy
from astropy.cosmology import Planck18
H0= Planck18.H0.value
Om0= Planck18.Om0
Ode0= Planck18.Ode0
def H(z):
    return np.sqrt(Om0*((1+z)**3) + Ode0)
def IntFunc(z):
    return 1/((1+z)*H(z))
def tL(z):
    integ, _ = scipy.integrate.quad(IntFunc, 0, z)
    tSec= integ/(H0/3.08e19)
    tGyr= (tSec)/(3.15e16)
    return tGyr
z=2
print("The lookbacktime in Gyr:", tL(z))
```

```python
#Question 4.21
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

bins=30
X= np.array(np.random.normal(0, 1, 10000))
Y= np.array(np.random.normal(2, 1, 10000))
x=np.linspace(-5, 7, 1000)
PDFX=norm.pdf(x, 0, 1)
PDFY=norm.pdf(x, 2, 1)
plt.hist(X, bins=30, density=True, color='skyblue', alpha=0.5, label='X', edgecolor='navy')
plt.plot(x, PDFX, label='PDF of X', linestyle=':', linewidth=3)
plt.hist(Y, bins=30, density=True, color='mintcream', alpha=0.5, label='Y', edgecolor='darkgreen')
plt.plot(x, PDFY, label='PDF of Y', linestyle='--', linewidth=3, color='green')
plt.annotate(r'$\mathcal{N}(0,1)$', color='blue', xy=(-1.75, 0.15), xytext=(-4, 0.15), arrowprops=dict(color='blue', arrowstyle='-|>', mutation_scale=25))
plt.annotate(r'$\mathcal{N}(2,1)$', color='green', xy=(3.75, 0.15), xytext=(4.5, 0.15), arrowprops=dict(color='green', arrowstyle='-|>', mutation_scale=25))
plt.xlabel('Values')
plt.ylabel('Density')
plt.grid(axis='y', linestyle='--')
plt.legend()
plt.show()


counts, xedges, yedges = np.histogram2d(X, Y, bins)
Xg, Yg = np.meshgrid((xedges[1:]+xedges[:-1])/2, (yedges[1:]+yedges[:-1])/2)

flat_counts = counts.flatten()
sorted_counts=np.sort(flat_counts)[::-1]

cumulative=np.cumsum(sorted_counts)
cumulative/= cumulative[-1]
sigma_levels=[0.68, 0.95, 0.997]
levels=[sorted_counts[np.searchsorted(cumulative, s)] for s in sigma_levels]
levels=sorted(levels)

plt.figure(figsize=(8,6))
img = plt.hist2d(X, Y, bins, cmap='Blues')
plt.colorbar(img[3])


contour=plt.contour(Xg, Yg, counts.T, levels=levels, colors='red')
plt.clabel(contour, fmt={levels[0]:'1σ', levels[1]:'2σ', levels[2]:'3σ'}, inline=True, fontsize=10)
plt.gca().set_aspect('equal', adjustable='box')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('2D histogram with Contour')
plt.grid(True, alpha=0.5, linestyle='--', color='black')
plt.show()
```
