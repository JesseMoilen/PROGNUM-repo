```python
#question 4.4
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
NH= np.random.normal(185, 8, 100000)

NHm = np.mean(NH)
NHstd = np.std(NH)
print(f'The mean of the randomly generated distribution = {NHm}')
print(f'The standard deviation of the randomly generated distribution = {NHstd}')
print('These values are quite similar to the given values of the mean and std when generating the distribution')

plt.title('The intensity of all the average heights (cm)')
plt.hist(NH, bins=40, color='blue', label='Average height intensity')
plt.plot([], [], color='white', label=f'The mean = {NHm:.4f}')
plt.plot([], [], color='white', label=f'The std = {NHstd:.4f}')
plt.axvline(NHm, color='green', label='sample mean', linewidth=3, alpha = 0.5)
plt.axvspan(NHm - NHstd, NHm + NHstd, label='std span', alpha=0.2, color='red')
plt.xlabel('Average height (cm)')
plt.ylabel('Intensity')
plt.legend()
plt.show()

h=190
P = norm.cdf(h, loc=185, scale=8)
print(f'The probability of someone having a height =< h = {P}')
top = 185+8
bot = 185-8

Pr = norm.cdf(top, loc=185, scale=8)-norm.cdf(bot, loc=185, scale=8)
print(f'Thus about {(Pr*100):.3f}% of values lie within one standard deviation')
```
